```python
token_nlp = spacy.load("tl_calamancy_trf")
sentence_nlp = spacy.load("en_core_web_trf")

@Language.component('boundaries')
def set_custom_boundaries(doc):
    for token in doc[0:6]: #Only the first n tokens
        if token.text == ")" or token.text == "—" or token.text == "–" or token.text == "-":
            doc[token.i+1].is_sent_start = True
    return doc

token_nlp.add_pipe('boundaries', first=True)

excel = pd.read_excel("Flooding articles.xlsx")
```

```python
%%time

# Check for all docs if it works

tagalog_tagged = []

for i in range(len(excel['Tagalog'])):

    doc = excel['Tagalog'][i]
    
    sentence_doc = sentence_nlp(doc.replace("\n", " "))
    
    sentences = []

    for sentence in sentence_doc.sents:
        sentences.append(sentence)
        
    tagged = ""

    for sent in sentences:
        
#         print(sent.text)

        if sent.text == "" or sent.text == " ":
            continue

        dashlist = ["–", "—", "-", ")"]
        dash_present = False
        
        for dash in dashlist:
            if sent.text[-1] == dash: # Specifically to deal with an issue of a dash at the END of a sentence, which breaks the tokenizer
                
                dash_present = True
                
                tokenized = token_nlp(sent.text[0:-1]) # Do not tokenize the dash

                tagged += "<s>\n"

                for token in tokenized:
                    if str(token) == "" or str(token) == " ":
                        continue
                        
                    if str(token) == "." and token.norm != tokenized[-1].norm:
                        tagged += str(token) + "\t" + str(token.ent_iob_) + "\n</s>\n<s>\n"
                        continue
                    
                    if str(token.ent_iob_) != "O":
                        tagged += str(token) + "\t" + str(token.ent_iob_) + "-" + str(token.ent_type_) + "\n"
                    else:
                        tagged += str(token) + "\t" + str(token.ent_iob_) + "\n"

                tagged += dash + "\tO\n" # Place the dash back at the end

                tagged += "</s>\n"

        if not dash_present:
            tokenized = token_nlp(sent.text)

            tagged += "<s>\n"

            for token in tokenized:
                if str(token) == "" or str(token) == " ":
                    continue
                
                if str(token) == "." and token.norm != tokenized[-1].norm:
                    tagged += str(token) + "\t" + str(token.ent_iob_) + "\n</s>\n<s>\n"
                    continue
                
                if str(token.ent_iob_) != "O":
                    tagged += str(token) + "\t" + str(token.ent_iob_) + "-" + str(token.ent_type_) + "\n"
                else:
                    tagged += str(token) + "\t" + str(token.ent_iob_) + "\n"
                    
            tagged += "</s>\n"
            
    tagalog_tagged.append(tagged)

print("No issues found!")
```

```python
excel['Tagalog tagged'] = tagalog_tagged

excel.to_excel("Flooding articles tagged.xlsx")
```
