```python
import openpyxl_patch
openpyxl_patch.patch_read_string_table()
import openpyxl
path = 'Flooding_test_data_09-04-24 marked.xlsx'
book = openpyxl.load_workbook(path)

# ws = book.get_sheet_by_name('Sheet1')
ws = book['Sheet1']

total_locations = []
total_streets = []
total_barangays = []
total_cities = []
total_rivers = []
total_buildings = []
total_bridges = []
total_provinces = []
total_municipalities = []
total_relevance = []
total_descriptive = []

for row in ws.iter_rows():
    if row[8].value == "English - Corrected":
        continue
    else:
        cell = row[8]

        article_text = ""
        for x in cell.value.r:
            article_text += x.t
        
        checklist = list(filter(None, article_text.split("\n"))) # List of all sentences in this article

        # One empty list for every sentence in the article
        article_locations = [[] for x in range(len(checklist))] # Always minimally one empty line, then you can minus index
        article_streets = [[] for x in range(len(checklist))]
        article_barangays = [[] for x in range(len(checklist))]
        article_cities = [[] for x in range(len(checklist))]
        article_rivers = [[] for x in range(len(checklist))]
        article_bridges = [[] for x in range(len(checklist))]
        article_buildings = [[] for x in range(len(checklist))]
        article_provinces = [[] for x in range(len(checklist))]
        article_municipalities = [[] for x in range(len(checklist))]
        article_relevance = [[] for x in range(len(checklist))]
        article_descriptive = [[] for x in range(len(checklist))]
        i = 0
        
        for x in cell.value.r:
            for batch in list(filter(None, x.t.split("\n"))):

                    # print(batch)
                    # print(checklist)
                
                    for j, line in enumerate(checklist):
                        if line == '':
                            continue
                        
                        if batch in line:
                            checklist[j] = line.removeprefix(batch)
                            i = j
                            break
        
            # Now you have the index, and you can assign the location to that
            if x.rPr:
                if x.rPr.color:
                    # print(x)
                    if x.rPr.b \
                    or x.rPr.i \
                    or x.rPr.color.rgb == 'FFFF0000' \
                    or x.rPr.color.rgb == 'FFFFC000' \
                    or x.rPr.color.rgb == 'FF92D050' \
                    or x.rPr.color.rgb == 'FF00B0F0' \
                    or x.rPr.color.rgb == 'FF7030A0' \
                    or x.rPr.color.rgb == 'FF00B050' \
                    or x.rPr.color.rgb == 'FF0070C0' \
                    or x.rPr.color.rgb == 'FF996600':

                        # print(x.t)
                        
                        article_locations[i].append(x.t)

                        # Relevance
                        if x.rPr.b:
                            article_relevance[i].append(1)
                        else:
                            article_relevance[i].append(0)

                        # Descriptive entities
                        if x.rPr.i:
                            article_descriptive[i].append(1)
                        else:
                            article_descriptive[i].append(0)
        
                        # Streets
                        if x.rPr.color.rgb == 'FFFF0000' and not x.rPr.i: 
                            article_streets[i].append(1)
                        else:
                            article_streets[i].append(0)
        
                        # Barangays
                        if x.rPr.color.rgb == 'FFFFC000' and not x.rPr.i: 
                            article_barangays[i].append(1)
                        else:
                            article_barangays[i].append(0)
        
                        # Cities
                        if x.rPr.color.rgb == 'FF92D050' and not x.rPr.i: 
                            article_cities[i].append(1)
                        else:
                            article_cities[i].append(0)
        
                        # Rivers
                        if x.rPr.color.rgb == 'FF00B0F0' and not x.rPr.i: 
                            article_rivers[i].append(1)
                        else:
                            article_rivers[i].append(0)
        
                        # Bridges
                        if x.rPr.color.rgb == 'FF7030A0' and not x.rPr.i: 
                            article_bridges[i].append(1)
                        else:
                            article_bridges[i].append(0)
        
                        # Buildings
                        if x.rPr.color.rgb == 'FF00B050' and not x.rPr.i: 
                            article_buildings[i].append(1)
                        else:
                            article_buildings[i].append(0)
        
                        # Provinces
                        if x.rPr.color.rgb == 'FF0070C0' and not x.rPr.i: 
                            article_provinces[i].append(1)
                        else:
                            article_provinces[i].append(0)
    
                        # Municipalities
                        if x.rPr.color.rgb == 'FF996600' and not x.rPr.i: 
                            article_municipalities[i].append(1)
                        else:
                            article_municipalities[i].append(0)
        
        total_locations.append(article_locations)
        total_streets.append(article_streets)
        total_barangays.append(article_barangays)
        total_cities.append(article_cities)
        total_rivers.append(article_rivers)
        total_buildings.append(article_buildings)
        total_bridges.append(article_bridges)
        total_provinces.append(article_provinces)
        total_municipalities.append(article_municipalities)
        total_relevance.append(article_relevance)
        total_descriptive.append(article_descriptive)

        # print(row)

print(total_streets)

# For all these locations, remove spaces at the start or end as a preprocessing step
for i, loc_list_list in enumerate(total_locations):
        for j, loc_list in enumerate(loc_list_list):
            if loc_list:
                for l, loc in enumerate(loc_list):
                    if loc:
                        total_locations[i][j][l] = loc.strip()

# Remove any stray empty strings or commas
for i, loc_list_list in enumerate(total_locations):
        for j, loc_list in enumerate(loc_list_list):
            total_locations[i][j] = list(filter(None, [x.replace(",", "") for x in loc_list]))
                        
print(total_locations)
# print(total_cities)

# Also print how many locations each category has
print("locations N:\t\t ", str(len([item for sublist in [item for sublist in total_locations for item in sublist] for item in sublist])))
print("streets N:\t\t ", str(sum([item for sublist in [item for sublist in total_streets for item in sublist] for item in sublist])))
print("barangays N:\t\t ", str(sum([item for sublist in [item for sublist in total_barangays for item in sublist] for item in sublist])))
print("municipalities N:\t ", str(sum([item for sublist in [item for sublist in total_municipalities for item in sublist] for item in sublist])))
print("cities N:\t\t ", str(sum([item for sublist in [item for sublist in total_cities for item in sublist] for item in sublist])))
print("rivers N:\t\t ", str(sum([item for sublist in [item for sublist in total_rivers for item in sublist] for item in sublist])))
print("buildings N:\t\t ", str(sum([item for sublist in [item for sublist in total_buildings for item in sublist] for item in sublist])))
print("bridges N:\t\t ", str(sum([item for sublist in [item for sublist in total_bridges for item in sublist] for item in sublist])))
print("provinces N:\t\t ", str(sum([item for sublist in [item for sublist in total_provinces for item in sublist] for item in sublist])))
print("relevance N:\t\t ", str(sum([item for sublist in [item for sublist in total_relevance for item in sublist] for item in sublist])))
print("descriptive N:\t\t ", str(sum([item for sublist in [item for sublist in total_descriptive for item in sublist] for item in sublist])))

```

```python
def load_pred_txtfile(filename:str) -> list:
    '''
    Loads the predictions in the target language from an external txt file
    
    Returns:
        list: the list of the predictions in the target language
    '''
    with open(filename, errors="ignore") as f:
        pred = f.readlines()

    for i in range(len(pred)):
            pred[i] = pred[i].replace("\n", "")

    # print(len(list(filter(None, pred))))
            
    return list(filter(None, pred))
```
